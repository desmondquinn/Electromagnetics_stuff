## with nanoparticle

```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt

sx = 0.8
sy = 0.8
sz = 0.8
cell = mp.Vector3(sx,sy,sz)
res = 300

wvl_min = 0.353
wvl_max = 0.355
wvl = (wvl_min+wvl_max)/2
fmin = 1/wvl_min
fmax = 1/wvl_max
fcen = 0.5*(fmin+fmax)
wvl_cen = 1/fcen
df = 0.2*fcen

dpml = 0.5*wvl_max

from meep.materials import Ag
bead_material = Ag

src_pt = mp.Vector3(0,0,0.5*sz-(1.1*dpml))
sources = [mp.Source(mp.GaussianSource(fcen,fwidth=df,is_integrated=True)
                         , component=mp.Ex
                         , center=src_pt, size=mp.Vector3(sx,sy,0) )]

geometry = [ 
        mp.Sphere(center=mp.Vector3(0,0,0), radius=0.017, material=Ag)
    ]

symmetries = [mp.Mirror(mp.Y)]

sim = mp.Simulation(cell_size=cell,
                    boundary_layers=[mp.PML(dpml)],
                    geometry=geometry,
                    sources=sources,
                    k_point=False,
                    eps_averaging=False,
                    symmetries=symmetries,
                    resolution=res) 

volume_plane1 = mp.Volume(size=mp.Vector3(sx,0,sz), center=mp.Vector3(0,0,0))
volume_plane2 = mp.Volume(size=mp.Vector3(sx,sy,0), center=mp.Vector3(0,0,0))
sim.plot2D(output_plane=volume_plane1)
sim.plot2D(output_plane=volume_plane2)

freq = [1/0.354]
transformed_fields_x = sim.add_dft_fields([mp.Ex],freq, 
                                        where = mp.Volume(center=mp.Vector3(0,0,0),size=mp.Vector3(0.1,0.1,0)))
transformed_fields_y = sim.add_dft_fields([mp.Ey],freq, 
                                        where = mp.Volume(center=mp.Vector3(0,0,0),size=mp.Vector3(0.1,0.1,0)))
transformed_fields_z = sim.add_dft_fields([mp.Ez],freq, 
                                        where = mp.Volume(center=mp.Vector3(0,0,0),size=mp.Vector3(0.1,0.1,0)))


sim.run(until_after_sources=mp.stop_when_fields_decayed(10, mp.Ex, mp.Vector3(0,0.035,0), 1e-4))

final_fields_x = sim.get_dft_array(transformed_fields_x,mp.Ex,0)
final_fields_y = sim.get_dft_array(transformed_fields_y,mp.Ey,0)
final_fields_z = sim.get_dft_array(transformed_fields_z,mp.Ez,0)
final_fields = np.sqrt((final_fields_x**2)+(final_fields_y**2)+(final_fields_z**2))
```

## Plotting Enhancement factor 

```python
#a = np.abs(final_fields)**2/np.abs(in_fields)**2
a = np.abs(final_fields)/np.abs(in_fields)
print(a.shape)
x = np.linspace(0,0.1,32)
y = np.linspace(0,0.1,32)
X, Y = np.meshgrid(x,y)
plt.contourf(X,Y,a,40)
plt.xlabel('X')
plt.ylabel('Y')
cbar = plt.colorbar()
cbar.set_label('Enhancement factor')
```
